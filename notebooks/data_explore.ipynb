```python
# Plot Particles 
def particle_display(df, bounds, no_of_particles):
    particles = df[(df['beamline_mass'] >= bounds[0]) & (df['beamline_mass'] <= bounds[1])]
    
    if particles.empty:
        print("No particles found in the given beamline_mass range.")
        return

    particles = particles[['run', 'subrun', 'beamline_mass', 'positionxtpc', 'positionztpc']]
    
    no_of_particles = min(len(particles), no_of_particles)
    
    if no_of_particles == 0:
        print("No particles to display.")
        return

    particles = particles.sample(n=no_of_particles).reset_index(drop=True)
    
    # First element always 0.0 needs to be removed.
    def remove_first_element(arr):
        """Removes the first element from the inner NumPy array while keeping the structure."""
        if isinstance(arr, np.ndarray) and len(arr) > 0 and isinstance(arr[0], np.ndarray):
            return np.array([arr[0][1:]])  # Keep it wrapped in an outer array
        return arr  # Return unchanged if not matching expected format

    # Apply to the target columns
    particles["positionxtpc"] = particles["positionxtpc"].apply(remove_first_element)
    particles["positionztpc"] = particles["positionztpc"].apply(remove_first_element)

    for i in range(no_of_particles):
        plt.figure(figsize=(5, 4))  # Create a new figure for each particle
        plt.scatter(particles['positionxtpc'][i][0], particles['positionztpc'][i][0], s=5, color='blue')
         
        plt.xlabel('X Position')
        plt.ylabel('Z Position')
        plt.title(f'Particle {i+1} Event Display - Run {particles["run"][i]}, Subrun {particles["subrun"][i]}, Mass {particles["beamline_mass"][i]} MeV')
        
        plt.show()  
```

```python
particle_display(df, [0, 2750], 1)
```

```python
# Function to remove first element from arrays
def remove_first_element(arr):
    """Removes the first element from the inner NumPy array while keeping the structure."""
    if isinstance(arr, np.ndarray) and len(arr) > 0 and isinstance(arr[0], np.ndarray):
        return arr[0][1:]  # Remove first value from the inner array
    return arr  # Return unchanged if not matching expected format

# Function to plot residualrange vs. trkdedx for each particle separately
def dEdx(df, bounds, no_of_particles):
    # Filter particles within beamline_mass bounds
    particles = df[(df['beamline_mass'] > bounds[0]) & (df['beamline_mass'] < bounds[1]) & (df['eventtype'] == 'pm')]

    if particles.empty:
        print("No particles found in the given beamline_mass range.")
        return

    # Keep only relevant columns
    particles = particles[['run', 'subrun', 'beamline_mass', 'residualrange', 'trkdedx']]
    
    # Limit the number of particles
    no_of_particles = min(len(particles), no_of_particles)
    
    if no_of_particles == 0:
        print("No particles to display.")
        return

    # Randomly sample the particles
    particles = particles.sample(n=no_of_particles).reset_index(drop=True)

    # Remove first element from residualrange & trkdedx arrays
    particles["residualrange"] = particles["residualrange"].apply(remove_first_element)
    particles["trkdedx"] = particles["trkdedx"].apply(remove_first_element)

    # Plot each particle separately
    for i in range(no_of_particles):
        plt.figure(figsize=(6, 5))  # Separate figure for each particle
        plt.scatter(particles['residualrange'][i], particles['trkdedx'][i], s=5, color='blue')
        
        plt.xlabel("Residual Range")
        plt.ylabel("TrkDedx")
        plt.title(f"Particle {i+1} - Run {particles['run'][i]}, Subrun {particles['subrun'][i]}, Mass {particles['beamline_mass'][i]} MeV")

        
        plt.show()
```

```python

dEdx(df, bounds=[1600, 2750], no_of_particles=5)
```
